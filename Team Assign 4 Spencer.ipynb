```python
# import regular expressions
import numpy as np
def steal_tweets(df,year):
    import re

# Set-up patterns to be removed for the tweets
    pat1 = r"http\S+"
    pat2 = r"#"
    pat3 = r"@"
    pat4 = r"FAV"
    pat5 = r"RE"
    pat6 = r"pic.\S+"
    combined_pat = r'|'.join((pat1, pat2, pat3, pat4, pat5, pat6))
# replace the patterns with an empty string
    df['stripped'] =  [re.sub(combined_pat, '', w) for w in df.text]

# might have double spaces now (because of empty string replacements above) - remove double empty spaces
    df['stripped']  = [re.sub(r"  ", ' ', w) for w in df.loc[:,'stripped']]
    df = df[np.where(df.stripped.str.contains('(?:^|\W)(theft|steal|stolen|thief)(?:$|\W)', flags = re.IGNORECASE), 1, 0)==1]
    return df[df['timestamp'].astype(str).str[0:4]==year]
```

```python
# import regular expressions
import numpy as np
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

def fees_tweets(df):
    import re

# Set-up patterns to be removed for the tweets
    pat1 = r"http\S+"
    pat2 = r"#"
    pat3 = r"@"
    pat4 = r"FAV"
    pat5 = r"RE"
    pat6 = r"pic.\S+"
    combined_pat = r'|'.join((pat1, pat2, pat3, pat4, pat5, pat6))

# replace the patterns with an empty string
    df['stripped'] =  [re.sub(combined_pat, '', w) for w in df.text]

# might have double spaces now (because of empty string replacements above) - remove double empty spaces
    df['stripped']  = [re.sub(r"  ", ' ', w) for w in df.loc[:,'stripped']]
    df = df[np.where(df.stripped.str.contains('(?:^|\W)(charges|fee|commission)(?:$|\W)', flags = re.IGNORECASE), 1, 0)==1]
    
# Instantiate the sentiment analyzer (if you haven't already done so)
    analyser = SentimentIntensityAnalyzer()

# Now get the compound sentiment score for each tweet
    df['C_Score'] = np.nan
    for index, row in df.iterrows(): 
        df.loc[index, 'C_Score'] = analyser.polarity_scores(row['stripped'])['compound']
    df['Sentiment'] = np.nan

# Loop through rows of dataframe and determine strings for new column "Sentiment"
    for index, row in df.iterrows(): 
        if df.loc[index, 'C_Score'] > 0.05 : 
            df.loc[index, 'Sentiment'] = "Positive"       
        elif df.loc[index, 'C_Score'] < -0.05 :
            df.loc[index, 'Sentiment'] = "Negative"   
        else : 
            df.loc[index, 'Sentiment'] = "Neutral"
    return df
# Let's take a look!
    #pd.set_option('display.max_colwidth', -1)
    #tweets_df[['stripped','C_Score']].head(-50)
```

```python
import numpy as np
def speed_tweets(df):
    import re

# Set-up patterns to be removed for the tweets
    pat1 = r"http\S+"
    pat2 = r"#"
    pat3 = r"@"
    pat4 = r"FAV"
    pat5 = r"RE"
    pat6 = r"pic.\S+"
    combined_pat = r'|'.join((pat1, pat2, pat3, pat4, pat5, pat6))

# replace the patterns with an empty string
    df['stripped'] =  [re.sub(combined_pat, '', w) for w in df.text]

# might have double spaces now (because of empty string replacements above) - remove double empty spaces
    df['stripped']  = [re.sub(r"  ", ' ', w) for w in df.loc[:,'stripped']]
    df= df[np.where(df.stripped.str.contains('(?:^|\W)(fast|quick|speed|slow|forever|timely|quickly|speedy|hurry)(?:$|\W)', flags = re.IGNORECASE), 1, 0)==1]
    analyser = SentimentIntensityAnalyzer()

# Now get the compound sentiment score for each tweet
    df['C_Score'] = np.nan
    for index, row in df.iterrows(): 
        df.loc[index, 'C_Score'] = analyser.polarity_scores(row['stripped'])['compound']
    df['Sentiment'] = np.nan

# Loop through rows of dataframe and determine strings for new column "Sentiment"
    for index, row in df.iterrows(): 
        if df.loc[index, 'C_Score'] > 0.05 : 
            df.loc[index, 'Sentiment'] = "Positive"       
        elif df.loc[index, 'C_Score'] < -0.05 :
            df.loc[index, 'Sentiment'] = "Negative"   
        else : 
            df.loc[index, 'Sentiment'] = "Neutral"
    return df
```

```python
# import regular expressions
import numpy as np
def undelivered(df,year):
    import re

# Set-up patterns to be removed for the tweets
    pat1 = r"http\S+"
    pat2 = r"#"
    pat3 = r"@"
    pat4 = r"FAV"
    pat5 = r"RE"
    pat6 = r"pic.\S+"
    combined_pat = r'|'.join((pat1, pat2, pat3, pat4, pat5, pat6))
# replace the patterns with an empty string
    df['stripped'] =  [re.sub(combined_pat, '', w) for w in df.text]

# might have double spaces now (because of empty string replacements above) - remove double empty spaces
    df['stripped']  = [re.sub(r"  ", ' ', w) for w in df.loc[:,'stripped']]
    df = df[np.where(df.stripped.str.contains('(?:^|\W)(showed up|missing|refund|reported|not delivered|never arrived|never delivered|not here|undelivered)(?:$|\W)', flags = re.IGNORECASE), 1, 0)==1]
    return df[df['timestamp'].astype(str).str[0:4]==year]
```
